# Preparing Models

```python
# Checking for duplicate rows
df.duplicated().sum()
```

```python
# droping duplicate rows
df.drop_duplicates(keep=False,inplace=True)
```

```python
# checking top 5 rows
df.head()
```

```python
# ## Create Hot Encoding for the categorical columns
Cp = pd.get_dummies(df["cp"], prefix='cp', prefix_sep='_')
Restecg = pd.get_dummies(df["restecg"], prefix = "restecg", prefix_sep="_")
Slope = pd.get_dummies(df["slope"],prefix = "slope" ,prefix_sep = "_")
Ca = pd.get_dummies(df["ca"],prefix = "ca" , prefix_sep ="_")
Thal = pd.get_dummies(df["thal"],prefix ="thal", prefix_sep ="_")
```

```python
# concatination of hot encoding and dataframe
df = pd.concat([df,Cp,Restecg,Slope,Ca,Thal],axis = 1)
```

```python
# checking the columns
df.columns
```

```python
# droping the following columns "cp","restecg","slope","ca","thal" 
df = df.drop(columns = ["cp","restecg","slope","ca","thal"], axis= 1)
```

# Train test split

```python
# Creating "X" feature and "y" target 
X = df.drop(columns ="target",axis = 1)
y = df["target"]
```

```python
# Importing train_test_split from sklearn package
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 440)
```

```python
# Importing logistic regression from sklearn package
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

# Standard Scaler

```python
# importind Standard scaler from sklearn package
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
```

```python
# fit the model
model.fit(X_train_std, y_train)
```

```python
X_test_std = scaler.transform(X_test)
```

```python
# importing accuracy_score from sklearn package
from sklearn.metrics import accuracy_score
```

```python
prediction = model.predict(X_test_std)
accuracy_score(y_test, prediction)
```

```python
# import classification report from sklearn package
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediction))
```

# Plot Decision Tree

```python
# Importing necessary libraries
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus
```

```python
# Creating decision tree
dot_data = StringIO()
export_graphviz(model, 
                out_file=dot_data,  
                filled=True, 
                rounded=True,
                special_characters=True,
                feature_names = X.columns,
                class_names=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('heart.png')
Image(graph.create_png())
```

```python
# checking for depth 5
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf.fit(X_train_std, y_train)

#Predict the response for train dataset
y_pred = clf.predict(X_train_std)

# Model Accuracy, how often is the classifier correct?
print("Train Accuracy:", accuracy_score(y_train, y_pred))
y_pred_test = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Test Accuracy:",accuracy_score(y_test, y_pred_test))
```

```python
# Checking for depth 5
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=5)

# Train Decision Tree Classifer
clf.fit(X_train_std, y_train)

#Predict the response for train dataset
y_pred = clf.predict(X_train_std)

# Model Accuracy, how often is the classifier correct?
print("Train Accuracy:", accuracy_score(y_train, y_pred))

y_pred_test = clf.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Test Accuracy:",accuracy_score(y_test, y_pred_test))
```
